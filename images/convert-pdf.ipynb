```python
from pdf2image import convert_from_path
from PIL import Image
import glob, os

DPI = 400  # key for sharp text

for pdf in glob.glob("*.pdf"):
    pages = convert_from_path(
    pdf,
    dpi=400,
)

    img = pages[0]

    # Convert to grayscale (cleaner text)
    img = img.convert("L")

    # Auto-crop margins (very important)
    bbox = img.getbbox()
    if bbox:
        img = img.crop(bbox)

    out = os.path.splitext(pdf)[0] + ".png"
    img.save(out, optimize=True)

    print(f"Saved {out}")
```
